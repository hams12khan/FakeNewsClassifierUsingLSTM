```python
import pandas as pd 
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM,Dense
from tensorflow.keras.preprocessing.text import one_hot
```

```python
train_data=pd.read_csv("train.csv")
```

```python
df=train_data.dropna()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x=df.drop("label",axis=1)
y=df["label"]
```

```python
x.head()
```

```python
y.head()
```

```python
print(x.shape)
print(y.shape)
```

```python
voc_size=5000
```

```python
message=x.copy()
```

```python
#this is because when we remove stop words then we
message.reset_index(inplace=True)
```

```python
message.head()
```

```python
from nltk.stem import PorterStemmer
import re 
```

```python
from nltk.corpus import stopwords
```

```python
#this is for removing all stopwords from the title
ps=PorterStemmer()
corpus=[]
for i in range(0,len(message)):
    review=re.sub("[^a-zA-Z]"," ",message["title"][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word not in stopwords.words("english")]
    review=" ".join(review)
    corpus.append(review)
```

```python
len(corpus)
```

```python
corpus[1]
```

```python
one_hot_vector=[one_hot(word,5000) for word in corpus]
```

```python
one_hot_vector
```

```python
#for equalization in size 
sen_length=20
embedded_docs=pad_sequences(one_hot_vector,padding="pre",maxlen=sen_length)
```

```python
embedded_docs
```

```python
embedded_docs[0]
```

```python
#creating the MOdel
embedding_vector_feature=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_feature,input_length=sen_length))
model.add(LSTM(100))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=["accuracy"])
```

```python
model.summary()
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.2,random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict(X_test)
```

```python
#here we are converting the output as 0 and 1  
y_pred=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics  import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
